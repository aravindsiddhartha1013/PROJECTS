```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with 7 more bitches pls!!!! .txt','r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2}, \d{1,2}:\d{2}\s[AP]M\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates = re.findall(pattern,data)
```

```python
dates
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})
# Replace '\u202f' with a regular space and remove the trailing ' - '
df['message_date'] = df['message_date'].str.replace('\u202f', ' ').str.rstrip(' - ')

# Converting the 'message_date' column to datetime format
df['message_date'] = pd.to_datetime(df['message_date'], format="%m/%d/%y, %I:%M %p")
```

```python
df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'^(.*?):\s', message, maxsplit=1)
    if len(entry) > 1:  # Check if there's a username and message
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

# Add to DataFrame
df['user'] = users
df['message'] = messages

# Drop the original 'user_message' column
df.drop(columns=['user_message'], inplace=True)

# Display the DataFrame
print(df)
```

```python
df['year'] = df['message_date'].dt.year
df['month'] = df['message_date'].dt.month_name()
df['Day'] = df['message_date'].dt.day
df['hour'] = df['message_date'].dt.hour
df['minute'] = df['message_date'].dt.minute
```
